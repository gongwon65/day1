# 빈도수 확인

```python
font_title = {'family':'batang','c':'b','size':20}
font_sub = {'family':'gulim','c':'k','size':10}

#--------------------------------------------------------------------

plt.figure(facecolor='white')
plt.hist2d(x, y, bins=(40,40), cmap=plt.cm.jet) #bins=(40,40) : 격자를 40x40로 나눈다.
cb = plt.colorbar()
cb.set_label('빈도에 따른 색 보정, bin=40', fontdict=font_sub)
plt.xlabel('X 좌표', fontdict=font_sub)
plt.ylabel('Y 좌표', fontdict=font_sub)
plt.gca().set_aspect('equal')
```
